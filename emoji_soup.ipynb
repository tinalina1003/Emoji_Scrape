```python
# import dependencies
from splinter import Browser
import requests
from bs4 import BeautifulSoup as bs
from pprint import pprint
    
```

```python
# create function to scrape
#session = requests.Session()

browser = Browser('chrome')
```

```python
url = 'https://carpedm20.github.io/emoji/'
#html = session.get(url)
browser.visit(url)
```

```python
html = browser.html
```

```python
soup = bs(html, 'html.parser')
soup
```

```python
table = soup.find('table')
table
```

```python
emoji_names = []

rows = table.find_all('tr', class_= 'en')[1:-1] # inconsistent elements throughout the whole list.. get rid of the ones we don't need

for emoji in rows:
    emoji_text =  ':' + emoji.text.split(':')[1] + ':'  # split into lists by using .split(':') and only the second element
    emoji_names.append(emoji_text) # add back in the : for emoji

emoji_names
```
